```python
x = int(input())

for i in range(x):
    num = input()
    t = 0
    c = 1
    for j in num:
        if j == '0':
            t += c
            c += 1
        else:
            c = 1
    print(t)
```

```python
import itertools, math

def minNum(n):
            x = [2,3,4,5,6,7,0]
            dic = {0:0,2:1,3:7,4:4,5:2,6:0,7:8}
            d =  math.ceil(n / 7)
            a = [p for p in itertools.product(x, repeat=d) if sum(p) == n]
            res = []
            for j in a:
                        t = 0
                        for i in range(d):
                                    t += dic[j[i]]*(10**i)
                        if t == 0:
                                    t = 6
                        res.append(t)
            res = filter(lambda x: len(str(x))== d, res)
            return min(res)

def maxNum(n):
            if n % 2 == 0:
                        res = "1"*(n//2)
            else:
                        res =  "7"+"1"*(n//2-1)
            return int(res)


def main():
            loop = int(input())
            for i in range(loop):
                        x = int(input())
                        print(minNum(x),maxNum(x))

__name__ == main()
```
