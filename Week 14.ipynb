```python
def gcd(x, y):
    while y:
        x, y = y, x %y
    return x

def lcm(x, y):
    res = (x * y) // gcd(x, y)
    return res

def main():
    loop = int(input())
    for i in range(loop):
        put = input().split()
        x = int(put[0])
        y = int(put[1])
        print(lcm(x, y))

__name__ = main()
```
