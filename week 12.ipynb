```python
x = int(input())

while x != 0:
    flag = 0
    for i in range(1, x):
        if x % i == 0:
            flag += 1
    if flag % 2 == 0:
        print('yes')
    else:
        print('no')

    x = int(input())
```

```python
def check(v, n, w):
    global x, flag
    if v == w:
        flag = True
        return
    for i in range(6, 0, -1):
        if x[i -1]:
            if (i + v) <= w:
                x[i -1] -= 1
                check(v + i, i, w)
                if flag:
                    break
                


case = 1
n = input()
while n != "0 0 0 0 0 0":
    flag = False
    x = n.split()
    x = list(map(int, x))

    numMax = 0

    for i in range(6):
        numMax += x[i] * (i + 1)

    if numMax % 2 != 0:
        print(f"Collection #{case}:")
        print("Can't be divided.")
    else:
        check(0, 6, numMax // 2)
        if flag:
            print(f"Collection #{case}:")
            print("Can be divided.")
        else:
            print(f"Collection #{case}:")
            print("Can't be divided.")    
    case += 1
    n = input()
```
